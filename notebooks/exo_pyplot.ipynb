# PYPLOT

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1 - Tracer un graphe linéaire des points (x, y) où:
- **x** : `[0, 1, 2, 3, 4]`  
- **y** : `[1, 4, 3, 8, 5]`

```python
# Fonction qui trace une courbe linéaire reliant des points x et y
def tracer_graphe_lineaire(x, y):
    # Trace une ligne reliant les points fournis
    plt.plot(x, y, marker='o', linestyle='-', color='red', label='y = f(x)')
    # Ajoute le titre et les labels des axes
    plt.title("Graphe linéaire de y = f(x)")
    plt.xlabel("x")
    plt.ylabel("y")
    # Affiche le graphique avec légende et grille
    plt.legend()
    plt.grid(True)
    plt.show()

# Exemple d'utilisation
tracer_graphe_lineaire(
    x=[0, 1, 2, 3, 4],
    y=[1, 4, 3, 8, 5]
)
```

## 2 - Tracer un histogramme à partir de données aléatoires générées par NumPy.

```python
# Fonction qui trace un histogramme à partir de n valeurs normales aléatoires
def tracer_histogramme(n):
    # Trace l'histogramme des valeurs aléatoires
    valeurs = np.random.randn(n)
    plt.hist(valeurs, bins=50, color='blue', edgecolor='black')
    # Ajoute le titre et les labels des axes
    plt.title("Histogramme de données aléatoires")
    plt.xlabel("Valeurs")
    plt.ylabel("Fréquence")
    # Affiche le graphique
    plt.show()

# Exemple d'utilisation
tracer_histogramme(10000)
```

## 3 - Tracer un diagramme à barres à partir des données suivantes : 
- **categories** : `['A', 'B', 'C', 'D', 'E']`  
- **valeurs** : `[25, 30, 15, 20, 10]`  

```python
# Fonction qui trace un diagramme à barres pour des catégories et des valeurs
def tracer_barres(categories, valeurs):
    # Trace les barres avec les valeurs affichées au-dessus
    bars = plt.bar(categories, valeurs, color='skyblue', edgecolor='black')
    plt.bar_label(bars, labels=[str(v) for v in valeurs], label_type='edge')
    # Ajoute le titre et les labels des axes
    plt.title("Diagramme à barres")
    plt.xlabel("Catégories")
    plt.ylabel("Valeurs")
    # Affiche le graphique
    plt.show()

# Exemple d'utilisation
tracer_barres(
    categories=['A', 'B', 'C', 'D', 'E'],
    valeurs=[25, 30, 15, 20, 10]
)
```

## 4 - Tracer un nuage de points à partir de données aléatoires générées par NumPy.

```python
# Fonction qui trace un nuage de points aléatoires
def tracer_nuage_points(n):
    # Génère n points aléatoires pour x et y
    x = np.random.rand(n)
    y = np.random.rand(n)
    # plt.scatter() trace des points
    plt.scatter(x, y, color='red', marker='o', alpha=1)
    # Ajoute le titre et les labels des axes
    plt.title("Nuage de points")
    plt.xlabel("x")
    plt.ylabel("y")
    # Affiche le graphique
    plt.show()

# Exemple d'utilisation
tracer_nuage_points(100)
```

## 5 - Tracer une courbe sinusoïdale en utilisant NumPy.

```python
# Fonction qui trace une courbe sinusoïdale de 0 à 2π
def tracer_sinus():
    # Trace sin(x) à partir de x
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x)
    plt.plot(x, y, color='blue', linewidth=2, label='sin(x)')
    # Ajoute le titre et les labels des axes
    plt.title("Courbe sinusoïdale")
    plt.xlabel("x")
    plt.ylabel("y")
    # Affiche les graduations de l'axe x en multiples de π
    ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
    labels = ['0', 'π/2', 'π', '3π/2', '2π']  # lisible et explicite
    plt.xticks(ticks, labels)
    # Affiche le graphique avec légende et grille
    plt.legend()
    plt.grid(True)
    plt.show()

# Exemple d'utilisation
tracer_sinus()
```

## 6 - Tracer deux courbes sur le même graphe à partir des données suivantes :
- **x** : `np.linspace(0, 10, 100)`  
- **y1** : `np.sin(x)`  
- **y2** : `np.cos(x)`

```python
# Fonction qui trace deux courbes sur le même graphique
def tracer_deux_courbes(x, y1_func, y2_func, label1='Courbe 1', label2='Courbe 2', titre="Deux courbes"):
    # Trace les deux courbes
    plt.plot(x, y1_func(x), label=label1, linewidth=2)
    plt.plot(x, y2_func(x), label=label2, linewidth=2)
    
    # Ajoute le titre et les labels des axes
    plt.title(titre)
    plt.xlabel("x")
    plt.ylabel("y")
    
    # Affiche le graphique avec légende et grille
    plt.legend()
    plt.grid(True)
    plt.show()

# Exemple d'utilisation
tracer_deux_courbes(
    x=np.linspace(0, 10, 100),
    y1_func=lambda x: np.sin(x),
    y2_func=lambda x: np.cos(x),
    label1='sin(x)',
    label2='cos(x)',
    titre="Courbes sin(x) et cos(x)"
)
```

## 7 - Tracer un diagramme en secteurs (camembert) à partir des données suivantes :
- **categories** : `['A', 'B', 'C', 'D']`  
- **valeurs** : `[30, 25, 15, 20]`

```python
# Fonction qui trace un diagramme en secteurs
def tracer_camembert(categories, valeurs):
    # Trace un camembert
    plt.pie(valeurs, labels=categories, autopct='%1.0f%%', startangle=90)
    # Ajoute le titre
    plt.title("Diagramme en secteurs")
    # Affiche le graphique
    plt.show()

# Exemple d'utilisation
tracer_camembert(
    categories=['A', 'B', 'C', 'D'],
    valeurs=[30, 25, 15, 20]
)
```

## 8 - Tracer une courbe polynomiale à partir des données suivantes :
- **x** : `np.linspace(-5, 5, 100)`  
- **y** : `3 * x**3 + 2 * x**2 - 8 * x + 5`

```python
# Fonction qui trace la courbe d'un polynôme donné par ses coefficients
def tracer_polynome(x, coefficients, label="y = f(x)"):
    # Trace la courbe du polynôme
    y = np.polyval(coefficients, x)
    plt.plot(x, y, label=label, color='purple', linewidth=2)
    # Ajoute le titre et les labels des axes
    plt.title("Courbe polynomiale")
    plt.xlabel("x")
    plt.ylabel("y")
    # Affiche le graphique avec légende et grille
    plt.legend()
    plt.grid(True)
    plt.show()

# Exemple d'utilisation
tracer_polynome(
    x=np.linspace(-5, 5, 100),
    coefficients=[3, 2, -8, 5],
    label="y = 3x^3 + 2x^2 - 8x + 5"
)
```

## 9 - Tracer un diagramme en barres empilées à partir des données suivantes :
- **categories** : `['A', 'B', 'C', 'D']`  
- **valeurs1** : `[10, 15, 20, 25]`  
- **valeurs2** : `[5, 10, 15, 20]`

```python
# Fonction qui trace un diagramme en barres empilées pour deux séries de valeurs
def tracer_barres_empilees(categories, valeurs1, valeurs2, label1="Groupe 1", label2="Groupe 2"):
    # Trace les barres empilées
    x = np.arange(len(categories))
    bars1 = plt.bar(x, valeurs1, label=label1, color='skyblue')
    bars2 = plt.bar(x, valeurs2, bottom=valeurs1, label=label2, color='orange')
    # Affiche les valeurs à l'intérieur des barres
    plt.bar_label(bars1, labels=[str(v) for v in valeurs1], label_type='center')
    plt.bar_label(bars2, labels=[str(v) for v in valeurs2], label_type='center')
    # Place les étiquettes des catégories sous chaque barre
    plt.xticks(x, categories)
    # Ajoute le titre et les labels des axes
    plt.title("Diagramme en barres empilées")
    plt.xlabel("Catégories")
    plt.ylabel("Valeurs")
    # Affiche le graphique avec légende
    plt.legend()
    plt.show()

# Exemple d'utilisation
tracer_barres_empilees(
    categories = ['A','B','C','D'],
    valeurs1 = [10,15,20,25],
    valeurs2 = [5,10,15,20],
    label1="Valeurs 1",
    label2="Valeurs 2"
)
```

## 10 -Tracer un graphique en boîte (boxplot) à partir des données suivantes :
- **data** :
  - Série 1 : `np.random.normal(0, 1, 100)`
  - Série 2 : `np.random.normal(2, 1, 100)`
  - Série 3 : `np.random.normal(-3, 2, 100)`

```python
# Fonction qui trace un boxplot pour plusieurs séries de données
def tracer_boxplot(series):
    # Récupère les noms et les données des séries
    labels = list(series.keys())
    data = list(series.values())
    # Trace un boxplot pour chaque série
    plt.boxplot(data, tick_labels=labels, patch_artist=True)
    # Ajoute le titre et le label de l'axe y
    plt.title("Graphique en boîte")
    plt.ylabel("Valeurs")
    # Affiche le graphique
    plt.show()

# Exemple d'utilisation
tracer_boxplot({
    "Série 1": np.random.normal(0, 1, 100),
    "Série 2": np.random.normal(2, 1, 100),
    "Série 3": np.random.normal(-3, 2, 100)
})
```
